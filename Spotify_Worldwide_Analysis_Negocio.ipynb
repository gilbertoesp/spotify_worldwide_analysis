### 1.1.- Descripcion del negocio.

Spotify es un servicio de musica en *streaming* desarroladdo por la compañia suiza *Spotify Technology*, cuyas oficinas centrales esta en Stockholm, Sweden. Ellos lanzaron sus servicios el 7 de Octubre 2008, Al dia de 20 de Agosto, 2018 el servicio esta presente en 65 regiones. Spotify es un servicio *freemium*; caracteristicas básicas son gratis con anuncios o limitaciones, mientras con cualidades adicionales, como calidad del streaming mejorada y descarga de musica, son ofrecidas via suscripciones pagadas.



### 1.2.- Objetivos del negocio.

Descubrir, manejar y compartir mas de 35 millones de canciones gratis, o actualizar a Spotify Premium para acceder a caracteristicas exclusivas incluyendo modo offline, mejoras en la calidad de sonido, Spotify Connect y libre de anuncios. Hoy, Spotify es el servicio de streaming de musica mas popular con una comunidad de 180 millones de usuarios, incluyendo 83 millones suscritores, a lo largo de 65 mercados.

A los largo de los años, Spotify se ha mantenido desarrollando cultura y continuamente adaptandose al mercado. Recientemente, ellos adquirieron Soundwave y Cord Project. ###  1.3.- Evaluación de la situación.
Como varias compañias de Internet, incluyendo _SnapChat_, _Twitter_, _Spotify_ no tiene muchas ganancias. Aun así sus ingresos no son pocos, tuvo $ \$4.9 $ mil millones en 2017, de acuerdo a sus declaraciones, con un porcentaje de crecimiento de $ 46 \% $ comparado con el año anterior. Todavía la compañia tiene una perdida de $ \$ 1.5 $ mil millones el año pasado.

Algunos también han tenido problemas con la forma que _Spotify_ licencia su música. Una demanda de $1.6$  mil millones de _Wixen Music Publishing_, por ejemplo, declarando que la compañia estaba usando miles de sus canciones ilegalmente.

_Spotify_ también ha incursionado en enviar artistas en giras basado en datos y algoritmos, abriendo una potencial fuente de ingreso. Organizando eventos con la información recabada por los usuarios, localización, artista, fechas; _Spotify_ podría ser la forma de mover la música. El año pasado, se creó _Rap Caviar_, un popular _playlist_ de _hip hop_ con mas de $8.7$ millones de seguidores. La _playlist_ se ha convertido en la forma en que esta generación descubre nueva música.

Derechos musicales son caros, y desde que solo hay unas cuantas firmas mayores. _Spotify_ tiene un poder de negociación limitado. Si no consigue un trato con ellos,  podrían perder un tercio de su librería y el interes de sus suscriptores. [CNBC - Spotify](https://www.cnbc.com/2018/02/28/spotify-ipo-business-model-flaws.html)

_Spotify_ empezó a ofecer, _"Sponsored Playlist"_, que son una oportunidad para que las marcas sean capaces de especificar a las audiencias lo que tienen en mente. Las marcas pueden crear sus propias _playlist_, permitiendoles alcanzar tipos especificos de consumidores realizando actividades especificas; ejemplo, una _playlist_ para hacer ejercico patrocinada por _Gatorade_ donde esta muestre anuncios cada cierto tiempo, claro para usuario no-pago. ###  1.4.- Objetivos de la mineria de datos

_Spotify_ guarda la música en sus propios servidores, pero usuarios individuales mantienen copias temporales de música recientemente reproducida en un archivo llamado _cache_. Si quieres volver a reproducir una canción, no es descargada de nuevo si no que es reproducida de forma local de este archivo. _Spotify_ también utiliza _social networking_ y busca por usuarios cerca que esten utilizando _Spotify_ para ver si sus _caches_ contienen la canción solicitada y permite a estos dispositivos proveer la música. De esta forma los servidores principales nunca esta sobrecargados y la música puede ser reproducidad 24/7 sin interrupciones.
